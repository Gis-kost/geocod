```python
# Установка необходимых пакетов
#!pip install geopy folium

from geopy.geocoders import Nominatim
import pandas as pd
from tqdm.notebook import tqdm_notebook
tqdm_notebook.pandas()
import folium

def geocode_address(address):
    """Получение координат адреса."""
    try:
        location = geolocator.geocode(address)
        if location is None:
            return {"latitude": None, "longitude": None}
        else:
            return {"latitude": location.latitude, "longitude": location.longitude}
    except Exception as e:
        print(f"Ошибка обработки адреса {address}: {e}")
        return {"latitude": None, "longitude": None}

# Объект гео-кодера
geolocator = Nominatim(user_agent="geo-coding")

# Список адресов
addresses = [
    'Москва, Красная площадь',
    'Санкт-Петербург, Невский проспект',
    'Новосибирск, Красный проспект'
]

# Преобразование списка адресов в DataFrame
df_addresses = pd.DataFrame(addresses, columns=['Address'])

# Применение функции геокодирования ко всему столбцу адресов
result_df = df_addresses.progress_apply(
    lambda row: pd.Series({**row.to_dict(), **geocode_address(row['Address'])}),
    axis=1
)

# Создание карты
m = folium.Map(location=[result_df["latitude"].mean(), result_df["longitude"].mean()],
               zoom_start=5,
               tiles='OpenStreetMap')

# Добавление маркеров на карту
for index, row in result_df.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=row['Address']).add_to(m)

# Отображение карты
display(m)
```
